```python
import numpy as np 
import matplotlib.pyplot as plt
import re

%matplotlib inline
```

## Training results

```python
file = open("model_soft_b6.log", "r")
#162
c = 0
lines = file.readlines()
scores = []
acc = []
eelbo = []
ppl = []

for line in lines:
    line = line.strip()
    name = line.strip().split(" ")
    if name[0] == 'Epoch':
#         print('!!')
        acc_, eelbo_, ppl_, xent_, kl_  = re.split('; [a-z]+: +', line)[1:]
        acc.append(float(acc_))
        eelbo.append(float(eelbo_))
        ppl.append(float(ppl_))


file.close()
```

```python
file = open("model_cat_sample_b6.log", "r")
#162
c = 0
lines = file.readlines()
scores = []
acc_reinf = []
eelbo_reinf = []
ppl_reinf = []

for line in lines:
    line = line.strip()
    name = line.strip().split(" ")
    if name[0] == 'Epoch':
#         print('!!')
        acc_, eelbo_, ppl_, xent_, kl_  = re.split('; [a-z]+: +', line)[1:]
        acc_reinf.append(float(acc_))
        eelbo_reinf.append(float(eelbo_))
        ppl_reinf.append(float(ppl_))


file.close()
```

```python
file = open("model_cat_enum_b6.log", "r")
#162
c = 0
lines = file.readlines()
scores = []
acc_exact = []
eelbo_exact = []
ppl_exact = []

for line in lines:
    line = line.strip()
    name = line.strip().split(" ")
    if name[0] == 'Epoch':
#         print('!!')
        acc_, eelbo_, ppl_, xent_, kl_  = re.split('; [a-z]+: +', line)[1:]
        acc_exact.append(float(acc_))
        eelbo_exact.append(float(eelbo_))
        ppl_exact.append(float(ppl_))


file.close()
```

```python
plt.plot(ppl, color='mediumseagreen', label='Soft Attn')
plt.plot(ppl_reinf, color='dodgerblue', label='REINFORCE')
plt.plot(ppl_exact, color='pink', label='Exact ELBO')
plt.yscale('log')
plt.title('PPL score during training')
plt.xlabel('iter')
plt.ylabel('score')
plt.legend();
plt.savefig('ppl_train.pdf')
```
